```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative
def f(x):
    return x**4 - 3*x**3 + 2

def df(x):
    return 4*x**3 - 9*x**2  
```

```python
# Initial value and settings
x_initial = 3  
learning_rate = 0.01  
epochs = 120  
batch_size = 10  
```

```python
# Different methods
```

```python
# Simple descending gradient: (SGD)
def gradient_descent(x, lr, epochs):
    x_values = [x]
    for _ in range(epochs):
        x = x - lr * df(x)
        x_values.append(x)
    return x_values
```

```python
# Mini-Batch Gradient Descent)
# In this example, only one random value is selected for each "minibatch".
def mini_batch_gradient_descent(x, lr, epochs, batch_size):
    x_values = [x]
    for _ in range(epochs):
        x = x - lr * df(x)
        x_values.append(x)
    return x_values
```

```python
# Decreasing gradient with momentum: (Momentum)
def momentum_gradient_descent(x, lr, epochs, momentum=0.9):
    x_values = [x]
    velocity = 0
    for _ in range(epochs):
        velocity = momentum * velocity - lr * df(x)
        x = x + velocity
        x_values.append(x)
    return x_values
```

```python
# Gradient descent with adaptive learning rate: (Adagrad)
def adagrad(x, lr, epochs):
    x_values = [x]
    cache = 0
    for _ in range(epochs):
        grad = df(x)
        cache += grad ** 2
        x = x - lr * grad / (np.sqrt(cache) + 1e-8)
        x_values.append(x)
    return x_values
```

```python
# Decreasing gradient: (Adam)
def adam(x, lr, epochs, beta1=0.9, beta2=0.999):
    x_values = [x]
    m, v = 0, 0
    t = 0
    for _ in range(epochs):
        t += 1
        grad = df(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * grad**2
        m_hat = m / (1 - beta1**t)
        v_hat = v / (1 - beta2**t)
        x = x - lr * m_hat / (np.sqrt(v_hat) + 1e-8)
        x_values.append(x)
    return x_values
```

```python
# Draw the graph of the function
x_range = np.linspace(-1, 3, 100)
y_range = f(x_range)

# Implementation of different methods
x = x_initial
sgd_path = gradient_descent(x, learning_rate, epochs)

x = x_initial
mini_batch_path = mini_batch_gradient_descent(x, learning_rate, epochs, batch_size)

x = x_initial
momentum_path = momentum_gradient_descent(x, learning_rate, epochs)

x = x_initial
adagrad_path = adagrad(x, learning_rate, epochs)

x = x_initial
adam_path = adam(x, learning_rate, epochs)

# Draw a separate graph for each algorithm

# Chart for SGD
plt.figure(figsize=(6, 4))
plt.plot(x_range, y_range, label="f(x) = x^4 - 3x^3 + 2")
plt.scatter(sgd_path, f(np.array(sgd_path)), color="red", s=15, label="SGD Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("SGD Optimization Path")
plt.legend()
plt.grid()
plt.show()

# Graph for Mini-Batch Gradient Descent
plt.figure(figsize=(6, 4))
plt.plot(x_range, y_range, label="f(x) = x^4 - 3x^3 + 2")
plt.scatter(mini_batch_path, f(np.array(mini_batch_path)), color="blue", s=15, label="Mini-Batch Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Mini-Batch Gradient Descent Optimization Path")
plt.legend()
plt.grid()
plt.show()

# Chart for Momentum
plt.figure(figsize=(6, 4))
plt.plot(x_range, y_range, label="f(x) = x^4 - 3x^3 + 2")
plt.scatter(momentum_path, f(np.array(momentum_path)), color="green", s=15, label="Momentum Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Momentum Gradient Descent Optimization Path")
plt.legend()
plt.grid()
plt.show()

# Diagram for Adagrad
plt.figure(figsize=(6, 4))
plt.plot(x_range, y_range, label="f(x) = x^4 - 3x^3 + 2")
plt.scatter(adagrad_path, f(np.array(adagrad_path)), color="orange", s=15, label="Adagrad Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Adagrad Optimization Path")
plt.legend()
plt.grid()
plt.show()

# Graph for Adam
plt.figure(figsize=(6, 4))
plt.plot(x_range, y_range, label="f(x) = x^4 - 3x^3 + 2")
plt.scatter(adam_path, f(np.array(adam_path)), color="purple", s=15, label="Adam Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Adam Optimization Path")
plt.legend()
plt.grid()
plt.show()

# Display the results
print(f"🔹 Optimal x with SGD: {sgd_path[-1]:.6f}")
print(f"🔹 Optimal x with Mini-Batch: {mini_batch_path[-1]:.6f}")
print(f"🔹 Optimal x with Momentum: {momentum_path[-1]:.6f}")
print(f"🔹 Optimal x with Adagrad: {adagrad_path[-1]:.6f}")
print(f"🔹 Optimal x with Adam: {adam_path[-1]:.6f}")
```
